1.FLIRT PREDICTION MODEL

A.Supervised Learning - Classification Statement

Import Necessary Libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

Load the Dataset

```python
df = pd.read_csv('flirt.csv')
df
```

Split the Data

```python
X = df["Chat"]
y = df["Label"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

B.Combination model: Countvector*MultinomialNB

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_train = count_vectorizer.fit_transform(X_train)
print(count_train)
```

```python
count_test = count_vectorizer.transform(X_test)
print(count_test)
```

```python
len(count_vectorizer.get_feature_names_out())
```

```python
print(count_train.toarray())
```

Train the Model

```python
clf = MultinomialNB()

clf.fit(count_train, y_train)
```

Predict on the test set

```python
pred = clf.predict(count_test)
pred
```

Check Accuracy Score

```python
from sklearn import metrics
score = metrics.accuracy_score(y_test, pred)
score
```

```python
print("accuracy:   %0.3f" % score)
```

Evaluate the Model

```python
cm = metrics.confusion_matrix(y_test, pred, labels=['non-flirt', 'non-flirt'])
cm
```

```python
from sklearn.metrics import classification_report

report=classification_report(y_test, pred)
print(report)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Clean up time column (strip extra spaces)
df['Time'] = df['Time'].str.strip()

# Convert 'Time' column to datetime format
try:
    df['Time'] = pd.to_datetime(df['Time'], format='%I:%M %p').dt.time
except ValueError:
    # If format fails, try infer datetime without format
    df['Time'] = pd.to_datetime(df['Time']).dt.time
```

```python
# Extract hour and minute
df['Hour'] = df['Time'].apply(lambda x: x.hour)
df['Minute'] = df['Time'].apply(lambda x: x.minute)
df
```

```python
# Count messages per hour
hourly_message_counts = df.groupby('Hour').size()
hourly_message_counts
```

```python
minute_message_counts = df.groupby(['Hour', 'Minute']).size()
minute_message_counts
```

```python
# Identify the most active hour
most_active_hour = hourly_message_counts.idxmax()
most_active_hour_count = hourly_message_counts.max()
print(f"Most Active Hour: {most_active_hour}:00 with {most_active_hour_count} messages")
```

```python
# Identify the most active minute within that hour
if most_active_hour in minute_message_counts.index.get_level_values(0):
    most_active_minute_in_hour = minute_message_counts[most_active_hour].idxmax()
    most_active_minute_count = minute_message_counts[most_active_hour].max()
    print(f"Most Active Minute in Hour {most_active_hour}:00: {most_active_minute_in_hour} minute with {most_active_minute_count} messages")
```
